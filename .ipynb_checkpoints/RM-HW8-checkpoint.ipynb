### Q2.1. Univariate comparison (3')

<font size='4'> 
    
- Suppose that we would like to examine the utility/effect of an intervention program for patients with PTSD.
- We measure PCL5 scores at completion (`pcl5week_score.completion`) and PCL5 score at 3-month follow-up (`pcl5month_score.3_month_follow_up`). Let's assume the first score is pre-intervention and the second score is post-intervention.
- Report the summary statistics for each variable including mean, std, median, Q1, and Q3.
- Note that each patient will receive such two PCL5 scores. Use a appropriate statistical test to perform the univariate comparison. Report the outputing statistic and p-value.
- Before you run the statistic test, determine the data type and check the missingness of two columns. In particular, report the number of NA values for each variable.

```python
# Write your own code
week_summary = ptsd_df["pcl5week_score.completion"].describe()
print()
month_summary = ptsd_df["pcl5month_score.3_month_follow_up"].describe()

print(week_summary)
print()
print(month_summary)
print()

# Generate a subset of these data with both sets of scores:
ptsd_scores = ptsd_df[["pcl5week_score.completion", "pcl5month_score.3_month_follow_up"]]
print(type(ptsd_scores))
print()

# Count the number of `NA` values in each column:
print(ptsd_scores.isna().sum())
ptsd_scores.describe()
print()

# Remove NA values before performing the statistical test:
ptsd_scores_paired = pd.DataFrame(ptsd_scores).dropna() 
print(ptsd_scores_paired.isna().sum())
print()

# Will perform a paired t-test (since pre- and post-intervention scores are measured for the same individual):
paired_ttest_scores = stats.ttest_rel(ptsd_scores_paired["pcl5week_score.completion"], 
                                      ptsd_scores_paired["pcl5month_score.3_month_follow_up"])
print(paired_ttest_scores) 

# There is a significant decrease between pcl5 scores pre- and post-intervention.
```

### Q2.2. Multiple Linear Regression (7')

<font size='4'>

- Select columns specified in the following code chunk and create a subset dataset named `ptsd_sub_df`.
- Fit a linear regression to examine the association between `caps_intake` (continuous outcome) and the remaining covariates (as predictors) using `ptsd_sub_df`.
    - Note that all covariates ending with `_code` are categorical variables.
- Use the instruction here to write the formula for linear regression in Python.
    - https://www.statsmodels.org/stable/example_formulas.html
- Report the output page including R2, adjusted R2, and parameter estimates, SE, 95% confidence intervals, and p-values.
- Provide a brief interpretation for all significant predictors (p<0.05) excluding the intercept.
- Relevant label information includes:
    - `employment_code`: 1: Employed, 2: Unemployed, 3: Retired, 4: Disabled/Unable to work, 5: Student, 6: Other.
    - `rank_code`: 1. Enlisted, 2: Officer, 3: Other

```python
# The following column names are used for linear regression.
# Do not delete.
relevant_col_names = ['caps_intake', 'age_iop', 'gender_code', 'sexualorient_code', 'race_code', 'ethnicity_code', 
                      'education_code', 'employment_code',
                      'rank_code', 'branch_code', 'mdd_code', 'ctq_total_score', 'sexual_trauma', 'sud_code']
```

```python
# Write your own code
ptsd_sub_df = ptsd_df[relevant_col_names]
ptsd_sub_df.head() 

ptsd_regression = smf.ols(formula='caps_intake ~ age_iop + C(gender_code) + C(sexualorient_code) + C(race_code) + C(ethnicity_code) + C(education_code) + C(employment_code) + C(rank_code) + C(branch_code) + C(mdd_code) + ctq_total_score + sexual_trauma + C(sud_code)', data=ptsd_sub_df)

ptsd_regression_output = ptsd_regression.fit() 

print(ptsd_regression_output.summary())
```

```python
# Write your interpretations below:
# The employment_code 4, rank_code 2, and ctq_total_score are all significantly associated with caps_intake. 

# The coefficient of ctq_total_score is 0.0417, suggesting a slight positive correlation between ctq_total_score and caps_intake score 
#        (as ctq_total_score increases, caps_intake is also likely to increase).

# The coefficient for rank_code 2 is -3.0559, and the employment_code 4 coefficient is 2.7408.   

# Rank code 2 represents military officers, and the negative magnitude of the coefficient suggests that having the rank of officer, 
#         on average, decreases an individual's caps_intake score.

# Employment code 4 designates individuals who are Disabled/Unable to work, and the larger positive magnitude 
#         of this coefficient suggests that having this designation as their employment code is likely to significantly 
#         increase an individual's caps_intake score. 

# Based on the multiple R-squared, at most 11.4% of the variation in caps_intake 
#        can be explained by the variation of the covariates included in the model.

# The adjusted R-squared suggests only 5.1% of the variation in caps_intake score can be explained by the variation in these covariates, 
#        however, which is consistent with most of the covariates in our model having no significant association with caps_intake score.
```
